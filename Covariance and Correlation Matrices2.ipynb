```python
import numpy as np
import pandas as pd

stocks = pd.read_excel("data-DJ30-july2011-june2013.xlsx")
stocks = stocks.sort_values(by = 'Date')
stocks = stocks.reset_index()
stocks = stocks.drop('index', axis = 1)
stocks
```

```python
column_names = stocks.columns.values.tolist()
for company in column_names[1:]:
    stocks[company+' Weekly Return'] = 0
    
    
for company in column_names[1:]:
    for i in range(1, 105):
        stocks[company+' Weekly Return'][i] = (stocks[company][i] - stocks[company][i- 1])/stocks[company][i- 1]
stocks = stocks.drop(index = 0)
stocks
```

```python
for company in column_names[1:]:
    stocks[company+' Mean Return'] = stocks.loc[:, company+' Weekly Return'].mean()
for company in column_names[1:]:
    stocks[company+' Demeaned'] = stocks[company+' Weekly Return'] - stocks[company+' Mean Return']
stocks
```

```python
#calculate the covariance matrix
T_bar = np.matrix(stocks.iloc[:, -8:])
cov_matrix = (1/103)*np.dot(np.transpose(T_bar), T_bar)
np.set_printoptions(suppress=True)
cov_matrix
```

```python
stocks2 = pd.read_excel("indeces-jul26-aug9-2012.xlsx")
stocks2
```

```python
column_names2 = stocks2.columns.values.tolist()
for company in column_names2[1:]:
    stocks2[company+' Daily Return'] = 0
    
    
for company in column_names2[1:]:
    for i in range(1, 11):
        stocks2[company+' Daily Return'][i] = np.log((stocks2[company][i]/stocks2[company][i- 1]))
stocks2 = stocks2.drop(index = 0)
stocks2
```

```python
for company in column_names2[1:]:
    stocks2[company+' Mean Return'] = stocks2.loc[:, company+' Daily Return'].mean()


for company in column_names2[1:]:
    stocks2[company+' Demeaned'] = stocks2[company+' Daily Return'] - stocks2[company+' Mean Return']
stocks2
```

```python
#covariance matrix of log returns
T_bar2 = np.matrix(stocks2.iloc[:, 10:13])
cov_matrix2 = (1/9)*np.dot(np.transpose(T_bar2), T_bar2)
np.set_printoptions(suppress=True)
cov_matrix2
```

```python
#covariance and correlation matrix 
indices = pd.read_excel("indices-july2011.xlsx")
indices
```

```python
log_return = indices.iloc[:, 1:].apply(lambda col: (col - col.shift(1))/col.shift(1))
log_return
```

```python
#mean normalized returns
mean_ret = log_return.apply(lambda col: col - col.mean()).dropna().values
mean_ret
```

```python
#covariance matrix
cov_indices = (1/142)*np.dot(np.transpose(mean_ret), mean_ret)
cov_indices
```

```python
#correlation matrix
log_return = log_return.dropna()
corr_ret = np.corrcoef(log_return, rowvar = False)
corr_ret
```
